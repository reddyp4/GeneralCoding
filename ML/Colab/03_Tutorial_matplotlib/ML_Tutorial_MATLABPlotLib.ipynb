**MAT Plot Lib**

```python
import matplotlib.pyplot as plt
```

```python
# Sample Data1
x = [1,2,3,4,5]
y = [2,4,6,8,10]

plt.plot(x,y)   #linear plot

plt.show()  #show the plot
```

```python
plt.scatter(x,y)  #scatter plot
```

```python
labels = ['A','B','C','D','E']
values = [3,7,5,2,8]
plt.bar(labels,values)    #bar plot
plt.show()    #show the plot
```

```python
plt.plot(x,y,linestyle = '--', color = 'red') #show dotted

plt.scatter(x,y,marker = 'x', color = 'green') #show dotted
```

```python
plt.plot(x,y,label='y vs. x')
plt.plot(y,x,label='x vs. y')
```

```python
import pandas as pd
```

```python
gas = pd.read_csv('gas_prices.csv')
```

```python
plt.figure(figsize = (8,5))
plt.title('Gas prices over time(USD)', fontdict = {'fontweight':'bold'})
plt.plot(gas['Year'],gas['USA'], 'b.-', label = 'United States')
plt.plot(gas.Year,gas.Australia, 'r.-', label = 'Australia')
plt.plot(gas.Year,gas.Canada, 'g.-', label = 'Canada')

plt.xlabel('Year')
plt.ylabel('USD')

plt.savefig('Gas_price_png.png', dpi = 300)
```

```python
fifa = pd.read_csv('fifa_data.csv')
```

```python
print(fifa.head(5)['Preferred Foot'].to_string()) #First five preferred foot
```

```python
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()

print(left)
print(right)
```

```python
left0 = left[0]
right0 = right[0]
```

```python
plt.pie([left0,right0],labels=['Left','Right'])
plt.title('Foot Preference of Players')
plt.show()  #removes any unwanted prints and prints just the plto
```
