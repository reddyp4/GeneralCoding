2. ABTesting:

A company is testing two different advertising strategies to see which results in more clicks on their website.

They randomly assign 5000 visitors to strategy A and 5000 visitors to strategy B.

The results show that 200 visitors from strategy A clicked on the website while 250 visitors from strategy B clicked on the website.

Test the hypothesis that the proportion of clicks is higher in strategy B than in strategy A using a significance level of 0.01.

Step1: State the null and alternative Hypothesis

Null(H0): Probability of strategy A = Probability of strategy B

Alternate(H1): pA < pB (since the number of clicks of B was higher)

Step2: Choose a significance level

That is alpha = 0.01

which means that we will 1% risk in making a Type-I error

Step3: Calculate the test statistic

Measure of how far the observed stastistic is from the hypothesised statistic

```python
'''
H0: PA=PB
H1: PA<PB

200 clicks in A
250 clicks in B

Sample variation not given, but pA is sample mean, can follow CLT

'''

#Sample number
nA = 5000
nB = 5000

#Probability
pA = 200/nA #sigma(xi)/n = sample mean, can follow CLT
pB = 250/nB

#z = ((phatA-phatB)-(pA-pB))/(sqrt( phatA*(1-phatA)/nA + phatB*(1-phatB)/nB ))

z =  (pA-pB)/sqrt( (pA)*(1-pA)/nA + (pB)*(1-pB)/nB )
print(z)
```
