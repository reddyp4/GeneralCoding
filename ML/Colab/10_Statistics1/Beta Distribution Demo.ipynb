- To generate Beta distribution in python
- Visualize how does Beta distrbution look like on varying shape parameters

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta
```

### Beta distribution using Numpy

```python
a, b = 80, 10

hist, bins = np.histogram(np.random.beta(a, b, 1000), bins=10, normed=True)
bin_centers = (bins[1:]+bins[:-1])*0.5
plt.plot(bin_centers, hist)
```

```python
a, b = 10, 100

hist, bins = np.histogram(np.random.beta(a, b, 1000), bins=10, normed=True)
bin_centers = (bins[1:]+bins[:-1])*0.5
plt.plot(bin_centers, hist)
```

#### Key takeaways

1. Numpy supports beta distribtion. Just plugin shape parameters a and b
2. Matplotlib can plot PDF using binned historgrams
