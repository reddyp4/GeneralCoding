Assume you have the following dataset

   25  37  24  28  2  35  22  31  53  41  64  29  120  72  2

a. Write a code in python to find

i. Mean

ii. Median

iii. Mode

iv. Standard deviation

v. Interquartile range

b. Can you find out which of the points are outliers in the above dataset?

c. Use any library in python to create boxplot of the above dataset

```python
#dataset
import numpy as np
import statistics as st
import seaborn as sns
import matplotlib.pyplot as plt


arr = np.array([25,37,24,28,2,35,22,31,53,41,64,29,120,72,2])
print(arr)

#Mean
print("Mean:",np.mean(arr))
#Median
print("Median:",np.median(arr))
#Mode
print("Mode:",st.mode(arr))
#standard dev
print("Standard Deviation:",np.std(arr))

#First Quartile
Q1 = np.median(arr[:int(len(arr)/2)])
Q3 = np.median(arr[int(len(arr)/2):])
print("IQR",Q3-Q1)

#Outliers by visualization, values over 80 are outliers
sns.boxplot(arr)
#Try with scatter plot
#plt.scatter(arr)
```

John is visiting planet Utopia and he figures out that the population is 1 million.

Utopians are very tall and their height varies between 180 cm and 250 cm.

John tries to estimate the population size by randomly identifying 10000 people n times and finds their mean heights.

Can you help John do the following using python:

(a) Generate population samples of size 10000 with height between 180 cm and 250 cm by varying n(number of times you identify 10000 random people) as following

i. 5

ii. 10

iii. 100

iv. 1000

v. 10000

(b) Plot histogram of the population means

(c) Does the population mean approach normal distribution as the sample size increases?

```python


#Population of 10000 people and their heights between 180 and 250 cm
def utopia(n):
  utopia_mean=[]
  for i in range(n):
    #Every time a new random population
    utopian_pop = np.random.random_sample(10000)*(250-180)+180
    utopia_mean.append(np.mean(utopian_pop))
  print(utopia_mean)
  plt.hist(utopia_mean)
  plt.show()

utopia(5)
```
